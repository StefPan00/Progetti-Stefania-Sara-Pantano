# Quesito 2b- Si chiede di trovare, per ogni continente la percentuale rispetto al totale mondiale del numero di casi

```python
#calcolo il numero totale di casi mondiali
totale_casi=df["new_cases"].sum()
print(f"Il numero totale dei casi mondiali sono stati:{totale_casi}") 

#calcolo la percentuale dei casi sui continenti
percentuale=filtro_casi_continenti / totale_casi * 100
percentuale= round(percentuale,2)

#calcolo la percentuale dei casi sui nan
filtro_nan = df[df['continent'].isna()]
percentuale_nan = filtro_nan["new_cases"].sum()/totale_casi *100
percentuale_nan= round(percentuale_nan,2)

#calcolo la percentuale totale
percentuale_totale = percentuale.sum() + percentuale_nan

#stampo a video le percentuali
print("Percentuale per continente:", percentuale)
print("Percentuale di NaN:", percentuale_nan)
print("Somma totale delle percentuali:", percentuale_totale) # per verifica che il totale sia 100%
```

# Quesito 3a- 3b: Selezionare i dati relativi all'Italia nel 2022 e mostrare con dei grafici adeguati l'evoluzione del casi totali dall'inizio alla fine dell'anno e il numero di nuovi casi rispetto alla data

```python
#Inserisco il formato data nella colonna date
df["date"] = pd.to_datetime(df["date"])

#Inserisco un filtro che seleziona Italy nella location e anno 2022 in date
filtroIt = (df["location"]=="Italy") & (df["date"].dt.year == 2022)
db3a=df[filtroIt]

#per visualizzare meglio i dati, raggruppo per mese del 2022
#inseriscono una colonna che chiamo mese, ed eseguo un group.by sommando i new_cases di quel mese
db3a.loc[:, "month"] = db3a["date"].dt.month
casi_mensili_2022 = db3a.groupby("month")["new_cases"].sum().reset_index()

#Creato il db a cui fare riferimento passo alla creazione dei grafici
sns.set_style("darkgrid")
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

db3a_filtrato = db3a[db3a["new_cases"] != 0] #escludo i valoti 0 
sns.lineplot(x=db3a_filtrato["date"], y=db3a_filtrato["new_cases"], ax=axs[0,0])
axs[0,0].set_title("Andamento giornaliero", fontweight="bold")
axs[0,0].set_ylabel("Nuovi casi")
axs[0,0].set_xlabel("Data")

sns.barplot(x="month", y="new_cases", data=casi_mensili_2022,ax=axs[0,1])
axs[0,1].set_title("Andamento mensile",fontweight="bold")
axs[0,1].set_ylabel("Nuovi casi")
axs[0,1].set_xlabel("Mese")

sns.lineplot(x="date", y="total_cases", data=db3a_filtrato, ax=axs[1,0])
axs[1,0].set_xlabel("Date")
axs[1,0].set_ylabel("Casi Totali")
axs[1,0].set_title("Somma cumulativa, casi totali", fontweight="bold")

fig.delaxes(axs[1, 1])

fig.suptitle("Analisi dei nuovi casi Covid-19 in Italia nel 2022", fontsize=16, fontweight="bold")
plt.show()
```

# Quesito 5- Riguardo le nazioni di Italia, Germania, Francia e Spagna in tutto il 2023: a. mostrare la somma dei pazienti ospitalizzati per ognuna b. se ci sono dati nulli, con un breve commento scrivere se può essere possibile gestirli tramite sostituzione o meno

```python
IGFS=(df["location"]=="Italy") | (df["location"]=="Germany") | (df["location"]=="France") | (df["location"]=="Spain") 
dfIGFS = df[IGFS]
filtro_2023 = dfIGFS["date"].dt.year == 2023
IGFS_2023 = dfIGFS[filtro_2023]
IGFS_2023 #db filtrato per il quesito 5

IGFS_2023_filtrato= IGFS_2023.loc[:,["location", "date", "hosp_patients","icu_patients", "weekly_hosp_admissions"]]

print("Shape IGF_2023", IGFS_2023_filtrato.shape)

# esaminiamo i valori nulli, facendo un group by per location
valori_nulli = IGFS_2023_filtrato.isnull().groupby(IGFS_2023_filtrato['location']).sum()

print("VALORI NULLI per colonna: \n", valori_nulli)
# hosp_patients presente il numero di ricoverati in osperdale giornalieri
# icu-patients sono il numero di persone in terapia intensiva, che quindi dovrebbero essere comprese nella hosp patients, in quanto non tutti i pz sono in icu
# weekly hosp admissions, tiene conto delle ospedalizzazioni registrate in una settimana
# #si analizza l'intero 2023 composto da 365 giorni, quindi 365 righe per ogni Nazione. Lo shape sarà 365x4
#si nota che la germania ha tutti i valori nulli nella colonna host_patients per il 2023. Mancate registrazioni
# per trattare questi valori si fa la somma dei pazienti ospedalizzati sulla colonna week, che ha il numero minore di valori nulli.
```

```python
# TRATTIAMO I VALORI NULLI DI WEEKLY:calcoliamo un indice che sarà dato dalla somma dei pz ospedalizzati (colonna weekly) / il numero di volte che il dato non nullo è presente.
# Questo ci permette di analizzare i dati in modo più oggettivio 

somma_pz = IGFS_2023_filtrato.groupby('location')['weekly_hosp_admissions'].sum()
count_nan = IGFS_2023_filtrato.groupby('location')['weekly_hosp_admissions'].count()
indice = round( (somma_pz / count_nan),2)

print(indice)
 #l'analisi delle ospedalizzazioni fornisce idea sullo stato degli ospadali dopo la campagna vaccinale
```

```python
""" il numero delle ospedalizzazioni settimanali risula in linea con le aspettative, dal momento nel 2023 era già in atto da un pò la campagna vaccinale
Per la gestione dei dati nulli, in questo caso si propone di trascurarli, non considerarli per l'analisi. In alternativa si potrebbe proporre un metodo
leggermente più sofisticato in cui i valori nulli vengano sostituiti con un valore medio tra il precedente e quello successivo"""
```

```python
indice.plot(kind='bar', color=["g", "mediumseagreen", "c", "skyblue"], figsize=(10, 6))

plt.title("Ospedalizzazioni Totali per Nazione", fontsize=16)
plt.xlabel("Nazione", fontsize=12)
plt.ylabel("Ospedalizzazioni Totali", fontsize=12)
plt.xticks(rotation=0)

#inseriamo il valore ottenuto sopra ogni barra 
for i, value in enumerate(indice):
    plt.text(i, value + max(indice) * 0.01, f'{value}', ha='center', fontsize=10)
    
plt.show()
```
